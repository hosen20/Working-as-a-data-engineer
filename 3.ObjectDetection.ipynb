```python
# importing necessary packages
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# reading the image
img = cv2.imread('/content/car_plate.jpeg')
```

```python
# defining a function to display the image
def display(img):
  # resilizing the display size
  fig = plt.figure(figsize=(5,5))
  # performs color correction
  new_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
  # displaying the image
  plt.imshow(new_img)

display(img)
```

```python
# russian plate charactersitics, which will help in finding the plate
plate_cascade = cv2.CascadeClassifier('/haarcascade_russian_plate_number.xml')
```

```python
# defining a function that draws a rectangle around the plate in the picture
def detect_plate(img):
  plate_img = img.copy()
  #finding the coordinates of the plate in the picture
  plate_rects = plate_cascade.detectMultiScale(plate_img,
    scaleFactor=1.3, minNeighbors=3)

  # drawing a rectangle around the plate
  for(x,y,w,h) in plate_rects:
    cv2.rectangle(plate_img, (x,y), (x+w, y+h), (0,0,255),3)
    return plate_img
```

```python
# getting the result of the image with the plate marked
result = detect_plate(img)
```

```python
# drawing the new image
display(result)
```
